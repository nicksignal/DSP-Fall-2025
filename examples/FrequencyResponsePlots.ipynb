```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MultipleLocator
```

## Defining the Transfer Function

This is a function that computes the transfer function for the right-sided exponential system that we covered in class. It has a pole at $z = 0.5$ and a zero at $z = 0$.

```python
def H(z):
    return z / (z - 0.5)
```

## Plotting Frequency Response Magnitude and Phase

Setting up the plotting variable `omega` (frequency) and evaluating the frequency response, $H(e^{i\omega})$

```python
L = 1001
omega = np.arange(L) * 2.0 * np.pi / L - np.pi
Hz = H(np.exp(1j * omega))
```

Plotting the magnitude, $|H(e^{i\omega})|$

```python
ax = plt.figure(figsize=(10,4))
plt.plot(omega, np.abs(Hz))
ax = plt.gca()
ax.set_facecolor((0.9,0.9,0.9))
plt.grid(color = 'white')
plt.xlabel('$\omega$')
plt.ylabel("$|H(e^{i\omega})|$")
plt.title("Magnitude of Frequency Response")
ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.1f}$\pi$'.format(val/np.pi) if (val !=0 and val != np.pi and val != -np.pi) else '-$\pi$' if val == -np.pi else '$\pi$' if val == np.pi else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi*0.2))
```

Plotting the phase, $\mathrm{Arg}(H(e^{i\omega}))$

```python
ax = plt.figure(figsize=(10,4))
plt.plot(omega, np.angle(Hz))
ax = plt.gca()
ax.set_facecolor((0.9,0.9,0.9))
plt.grid(color = 'white')
plt.xlabel('$\omega$')
plt.ylabel("$\mathrm{Arg}(H(e^{i\omega}))$")
plt.title("Phase of Frequency Response")
ax.xaxis.set_major_formatter(FuncFormatter(
   lambda val,pos: '{:.1f}$\pi$'.format(val/np.pi) if (val !=0 and val != np.pi and val != -np.pi) else '-$\pi$' if val == -np.pi else '$\pi$' if val == np.pi else '0'
))
ax.xaxis.set_major_locator(MultipleLocator(base=np.pi*0.2))
```
